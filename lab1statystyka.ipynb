Ściągnij plik csv z https://extranet.who.int/tme/generateCSV.asp?ds=mdr_estimates. Zaimportuj wybraną kolumnę numeryczną z pliku do Pythona. Oblicz podstawowe statystyki dla zaimportowanych danych.

```python
import pandas as pd
import numpy as np
df = pd.read_csv('MDR_RR_TB_burden_estimates_2022-12-14.csv')
col = df['e_rr_pct_new']
col.describe()
```

```python
mean = np.mean(col)
print("Średnia: " + str(mean))
std = np.std(col)
print("Odchylenie standardowe: " + str(std))
median = np.median(col)
print("Mediana: " + str(median))
min = np.min(col)
print("Minimum: " + str(min))
max = np.max(col)
print("Maximum: " + str(max))
percentile25 = np.percentile(col, 25)
print("Percentyl dla 250%: " + str(percentile25))
percentile50 = np.percentile(col, 50)
print("Percentyl dla 50%: " + str(percentile50))
percentile75 = np.percentile(col, 75)
print("Percentyl dla 75%: " + str(percentile75))
```

Wgraj plik Wzrost.csv. Użyj dla niego funkcji z modułu statistics. Czym różnią się funkcje dotyczące wariancji i odchylenia standardowego?

```python
import statistics
df = pd.read_csv('Wzrost.csv')
tmp = list(df)
l = [float(i) for i in tmp]
variance = statistics.variance(l)
stdev = statistics.stdev(l)
print("Wariancja: " + str(variance))
print("Odchylenie standardowe: " + str(stdev))

# Funkcja variance() zwraca wartość wariancji, czyli średni kwadrat różnicy wartości poszczególnych elementów zbioru danych od ich średniej.
# Natomiast funkcja stdev() zwraca wartość odchylenia standardowego, czyli pierwiastek kwadratowy wariancji.
```

Znajdź dowolny plik csv. Zastosuj do niego statystyki z biblioteki scipy.stats. Sprawdź, jakie inne statystyki opisowe można znaleźć w tej bibliotece.

```python
from scipy import stats
df = pd.read_csv('bank.csv', sep=';')
mean = stats.describe(df['balance']).mean
print("Średnia: " + str(mean))
mode = stats.mode(df['balance'])
print("Wartość najczęściej występująca: " + str(mode))
median = stats.median_abs_deviation(df['balance'])
print("Mediana: " + str(median))
variation = stats.variation(df['balance'])
print("Wariancja: " + str(variation))
skewness = stats.skew(df['balance'])
print("Skośność danych: " + str(skewness))
kurtosis = stats.kurtosis(df['balance'])
print("Kurtoza danych: " + str(kurtosis))
```
